### Fonction à une variable

```python
def logistique(x,K,R,S):
    return K/(1+R*np.exp(-S*x))
```

```python
x=np.linspace(-10,10,1000)
y1=logistique(x,1,1,1)
y2=logistique(x,1,10,1)
y3=logistique(x,3,1,1)
y4=logistique(x,1,1,20)
y5=logistique(x,3,2,10)
```

```python
plt.plot(x,y1,label="simple sigmoide")
plt.plot(x,y2,label="triplet (K,R, s)= (1,10,1)")
plt.plot(x,y3,label="triplet (K,R, s)=  (3,1,1)")
plt.plot(x,y4,label="triplet (K,R, s)=  (1,1,20)")
plt.plot(x,y5,label="triplet (K,R, s)=  (3,2,10)")
plt.title(label="Fonction logistique:\n f(x,K,R,S)= $\dfrac{R}{1+K{e}^{-Sx}}$")
plt.grid()
txt=" chart by @Priestley"
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=12)
plt.legend(loc=6)
```

###  Fonction à deux  variables

#### La courbe en cloche  f(x,y)=${e}^{\left(-{x}^{2}-{y}^{2}\right)}$

```python
def gauss(x,y):
    return 10*np.exp(-x**2-y**2)
x=np.linspace(-2,2,50)
y=np.linspace(-2,2,50)
```

```python
X,Y=np.meshgrid(x,y)
z=gauss(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(40, -30)
ax.set_title(r" La gaussienne à 3 dimensions :${e}^{\left(-{x}^{2}-{y}^{2}\right)}$")
ax.plot_surface(X,Y,z,rstride=10, cstride=10, color='orangered', edgecolors='yellow', lw=0.6)
plt.show()
```

### Fonction: g(x,y)= $\dfrac{x^2}{169}-\dfrac{y^2}{144}$

```python
def g(x,y):
    return (x**2/169)-(y**2/144)
x=np.linspace(-2,2,50)
y=np.linspace(-2,2,50)
```

```python
X,Y=np.meshgrid(x,y)
z=g(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(40, -30)
ax.plot_surface(X,Y,z,rstride=10, cstride=10, color='orangered', edgecolors='yellow', lw=0.6)
plt.show()
```

### fonction h(x,y)= $\dfrac{x^2}{169}+\dfrac{y^2}{144}$

```python
def h(x,y):
    return (x**2/169)+(y**2/144)
x=np.linspace(-2,2,50)
y=np.linspace(-2,2,50)
```

```python
X,Y=np.meshgrid(x,y)
z=h(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(40, -30)
ax.plot_surface(X,Y,z,rstride=10, cstride=10, color='orangered', edgecolors='yellow', lw=0.6)
plt.show()
```

### 	l(x,y)=$\mathrm{sin}\left(x-y+{x}^{3}\right)+\mathrm{arccos}\left(\frac{x}{{x^3}+{3}}\right)$

```python
def l(x,y):
    return np.sin(x-y+x**3)+np.arccos(((x/(x**2)+3)))
x= np.linspace(-np.pi,np.pi,500)
y= np.linspace(-np.pi,np.pi,500)
```

```python
X,Y=np.meshgrid(x,y)
z=l(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(40, -30)
ax.plot_surface(X,Y,z,rstride=10, cstride=10, color='orangered', edgecolors='yellow', lw=0.6)
plt.show()
```
